```python
import requests
```

```python
API_KEY = ''
```

```python
league_id= 39
season = 2023
endpoint = '/players'
query = f"{endpoint}?league={league_id}&season={season}"
url = f"https://v3.football.api-sports.io/{query}"

headers = {
  'x-rapidapi-key': API_KEY,
  'x-rapidapi-host': 'v3.football.api-sports.io'
}

response = requests.get(url, headers=headers)

response.json()
```

```python
total_pages = response.json()['paging']['total']
player_list = response.json()['response']
```
